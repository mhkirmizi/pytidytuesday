```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-07-01/weekly_gas_prices.csv')
```

```python
df.head()
```

```python
df = df.query('formulation == "all" and fuel == "gasoline" and grade == "regular"')
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['week'] = df['date'].dt.isocalendar().week
```

```python
df = df.query('year not in [1990, 1991,1992,2025] and week != "53"')
```

```python
df = df[df['week'] != 53]
```

```python
df.head()
```

```python
fig = px.density_heatmap(df, x = 'week', y= 'year', 
                         z = 'price', 
                         nbinsx=52, 
                         nbinsy= 30,
                         histfunc='avg', 
                          width=1920, height=1080,
                         )
fig.update_layout(
    title=dict(
        text="Weekly Gas Price in US", 
        font=dict(size=54, color="black", family="Roboto", weight='bold'), 
        x=0.5, 
        xanchor='center'
    ), 
    xaxis=dict(
        title=dict(
            text=""),
        tickfont=dict(family="Roboto", size=20, color="black")
    ),
    yaxis=dict(
        title=dict(
            text=""),
        tickfont=dict(family="Roboto", size=20, color="black"),
    ), 
    legend_title='Species',  
    margin=dict(t=90, b=50) 
)
fig.show()
fig.write_image("Weekly_gas_prices.png", width=1920, height=1080, scale=2)
```
